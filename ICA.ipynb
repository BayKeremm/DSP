# **Blind Source Seperation**
Alken Rrokaj

The BSS model assumes the existence of $s_1(t),... ,s_n(t)$ independent signals and the observation of as many mixtures $x_1(t),...,x_n(t)$ these mixtures being linear and instantaneous, i.e., $x_i(t) = \sum_{j=1}^{n} a_{ij}s_j(t)$ for each $i=1,n$ (Cardoso, J-F., 1998).\
The mixing equation:
\begin{align*}
x(t) & = As(t) \\
\end{align*}

```python
%matplotlib inline
import librosa
import librosa.display
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
import scipy
import soundfile as sf
from scipy.io import wavfile
```

### **Reading the Files**

```python
# samplerate, sr, 48kHz on both of the files
click_mix, sr = sf.read('./audio/ClickCloserMix.wav')
whislte_mix, sr = sf.read('./audio/WhistleCloserMix.wav')
```

ClickCloserMix.wav

```python
# Waveplot
plt.figure(figsize=(14, 5))
librosa.display.waveshow(click_mix, sr=sr)

# Spectogram
click_mix_stft = librosa.stft(click_mix)
click_mix_stft_db = librosa.amplitude_to_db(abs(click_mix_stft))
plt.figure(figsize=(14, 5))
librosa.display.specshow(click_mix_stft_db, sr=sr, x_axis='time', y_axis='hz')

# Audio
ipd.Audio(click_mix,rate=sr)
```

WhistleCloserMix.wav

```python
# Waveplot
plt.figure(figsize=(14, 5))
librosa.display.waveshow(whislte_mix, sr=sr)

# Spectogram
whislte_mix_stft = librosa.stft(whislte_mix)
whislte_mix_stft_db = librosa.amplitude_to_db(abs(whislte_mix_stft))
plt.figure(figsize=(14, 5))
librosa.display.specshow(whislte_mix_stft_db, sr=sr, x_axis='time', y_axis='hz')

# Audio
ipd.Audio(whislte_mix,rate=sr)
```

### **Syncing the Files**

From [allisonnicoledeal/VideoSync](https://github.com/allisonnicoledeal/VideoSync) on GitHub
1. Fourier Transform of Audio Signal: Split audio into bins and apply the Fourier transform on each bin using the numpy library
2. The Fourier transform converts each bin data from the time domain to the frequency domain.
3. Peak Frequencies: Identify the frequency with the highest intensity in each bin to create a peak frequency constellation.
4. Frequency Constellation Alignment: Determine time offset by aligning frequency constellations of the two audio files.
